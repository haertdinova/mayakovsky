```python
import re
from collections import Counter
from pathlib import Path
import pandas as pd
```

```python
df = pd.read_excel('lines.xlsx')
df
```

```python
for i in range(0, len(df)):
  if len(df['metre'][i]) == 3:
    df.loc[i, 'metre'] = df['metre'][i][0] + df['metre'][i][1]
  else:
    df.loc[i, 'metre'] = df['metre'][i][0] + df['metre'][i][1] + df['metre'][i][2]
df
```

```python
#Усредненные профили ударности для всех метров
metres = []
number_of_lines = []
for i in range(0, len(df)):
  current_metre = df['metre'][i]
  if current_metre not in metres:
    metres.append(current_metre)
    syllable_order = []
    for n in range(0, len(df)):
      if df['metre'][n] == current_metre:
        current_line = df['line'][n]
        with open(f'lines/{current_metre}.txt', 'a', encoding='utf-8') as file:
          file.write(current_line + '\n')
          file.close()
        vowels_in_line = re.findall(r'[АИЭУЕЁЫОЮЯаиэуеёыоюя]<?', current_line)
        for idx, accent in enumerate(vowels_in_line):
            right_idx = idx + 1
            if accent.endswith('<'):
                syllable_order.append(right_idx)
      else:
        continue
    with open(f'lines/accents/accents_{current_metre}.txt', 'w', encoding='utf-8') as file:
          file.write(str(syllable_order))
          file.close()
    line_count = sum(1 for line in open(f'lines/{current_metre}.txt', encoding='utf-8'))
    number_of_lines.append(line_count)
    accent_count = Counter(syllable_order)
    accents = []
    accents_percentage = []
    for key, value in accent_count.items():
      value_percent = value / line_count
      accents.append(str(key))
      accents_percentage.append(value_percent)
    for_accents_df = {'accent': accents, 'percent': accents_percentage}
    accents_df = pd.DataFrame(for_accents_df)
    accents_df.to_excel(f'lines/stats/{current_metre}.xlsx')
  else:
    continue
for_number_of_lines_df = {'metre': metres, 'number_of_lines': number_of_lines}
lines_total = pd.DataFrame(for_number_of_lines_df)
lines_total.to_excel('lines_total.xlsx')
```

```python
#Профили ударности по годам
metres = []
years = [] 
number_of_lines = []
for i in range(0, len(df)):
    current_metre = df['metre'][i]
    if current_metre not in metres:
        metres.append(current_metre)
        years_in_metre = []
        for n in range(0, len(df)):
            if df['metre'][n] == current_metre:
                year = df['year'][n]
                if year not in years_in_metre:
                    years_in_metre.append(year)
        for year in years_in_metre:
            syllable_order = []
            year_line_count = 0
            for n in range(0, len(df)):
                if df['metre'][n] == current_metre and df['year'][n] == year:
                    current_line = df['line'][n]
                    with open(f'lines_by_years/{current_metre}_{year}.txt', 'a', encoding='utf-8') as file:
                        file.write(current_line + '\n')
                    vowels_in_line = re.findall(r'[АИЭУЕЁЫОЮЯаиэуеёыоюя]<?', current_line)
                    for idx, accent in enumerate(vowels_in_line):
                        right_idx = idx + 1
                        if accent.endswith('<'):
                            syllable_order.append(right_idx)
                    year_line_count += 1
            with open(f'lines_by_years/accents/accents_{current_metre}_{year}.txt', 'w', encoding='utf-8') as file:
                file.write(str(syllable_order))
            if year_line_count > 0:
                accent_count = Counter(syllable_order)
                accents = []
                accents_percentage = []
                for key, value in accent_count.items():
                    value_percent = value / year_line_count
                    accents.append(str(key))
                    accents_percentage.append(value_percent)
                for_accents_df = {'accent': accents, 'percent': accents_percentage}
                accents_df = pd.DataFrame(for_accents_df)
                accents_df.to_excel(f'lines_by_years/stats/{current_metre}_{year}.xlsx')
            years.append(year)
            number_of_lines.append(year_line_count)
            if 'final_metres' not in locals():
                final_metres = []
            final_metres.append(current_metre)
    else:
        continue
```

```python
#Профили ударности по текстам
metres = []
poems = [] 
number_of_lines = []
for i in range(0, len(df)):
    current_metre = df['metre'][i]
    if current_metre not in metres:
        metres.append(current_metre)
        poems_in_metre = []
        for n in range(0, len(df)):
            if df['metre'][n] == current_metre:
                poem = df['name'][n]
                if poem not in poems_in_metre:
                    poems_in_metre.append(poem)
        for poem in poems_in_metre:
            syllable_order = []
            poem_line_count = 0
            for_file_name = poem[:-3]
            for n in range(0, len(df)):
                if df['metre'][n] == current_metre and df['name'][n] == poem:
                    current_line = df['line'][n]
                    with open(f'lines_in_poems/{current_metre}_{for_file_name}.txt', 'a', encoding='utf-8') as file:
                        file.write(current_line + '\n')
                    vowels_in_line = re.findall(r'[АИЭУЕЁЫОЮЯаиэуеёыоюя]<?', current_line)
                    for idx, accent in enumerate(vowels_in_line):
                        right_idx = idx + 1
                        if accent.endswith('<'):
                            syllable_order.append(right_idx)
                    poem_line_count += 1
            with open(f'lines_in_poems/accents/accents_{current_metre}_{for_file_name}.txt', 'w', encoding='utf-8') as file:
                file.write(str(syllable_order))
            if poem_line_count > 0:
                accent_count = Counter(syllable_order)
                accents = []
                accents_percentage = []
                for key, value in accent_count.items():
                    value_percent = value / poem_line_count
                    accents.append(str(key))
                    accents_percentage.append(value_percent)
                for_accents_df = {'accent': accents, 'percent': accents_percentage}
                accents_df = pd.DataFrame(for_accents_df)
                accents_df.to_excel(f'lines_in_poems/stats/{current_metre}_{for_file_name}.xlsx')
            poems.append(poem)
            number_of_lines.append(poem_line_count)
            if 'final_metres' not in locals():
                final_metres = []
            final_metres.append(current_metre)
    else:
        continue
```
